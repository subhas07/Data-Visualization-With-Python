```python
import seaborn as sns
```

```python
#We use distribution plots to visualize the distribution of quantitative data.

#Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

#distplot/histplot
#jointplot
#pairplot
#rugplot
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'],bins=1)
```

```python
sns.displot(df['total_bill'],bins=3)
```

```python
sns.displot(df['total_bill'],bins=10)
```

```python
sns.displot(df['total_bill'],bins=15)
```

```python
sns.displot(df['total_bill'],bins=100,kde=True)
```

```python
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(2,2,4)
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hist')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='resid')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

```python
sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```
